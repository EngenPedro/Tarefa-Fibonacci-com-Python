```python
print('Seja Bem-Vindo(a) a Sequencia de Fibonacci!')
print('-'*30)
n1 = int(input('Quantos termos você deseja ver ?'))
print('-'*30)
t1 = 0
t2 = 1
print('{} > {}'.format(t1, t2), end='')
cont = 3
while cont <= n1:
  t3 = t1 + t2
  print(' > {}'.format(t3), end='')
  t1 = t2
  t2 = t3
  cont += 1
```
